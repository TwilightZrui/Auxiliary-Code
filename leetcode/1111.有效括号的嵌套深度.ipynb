#### 1. 问题描述

**贪心算法**

**栈**

有效括号字符串 仅由 "(" 和 ")" 构成，并符合下述几个条件之一：

空字符串
连接，可以记作 AB（A 与 B 连接），其中 A 和 B 都是有效括号字符串
嵌套，可以记作 (A)，其中 A 是有效括号字符串
类似地，我们可以定义任意有效括号字符串 s 的 嵌套深度 depth(S)：

s 为空时，depth("") = 0
s 为 A 与 B 连接时，depth(A + B) = max(depth(A), depth(B))，其中 A 和 B 都是有效括号字符串
s 为嵌套情况，depth("(" + A + ")") = 1 + depth(A)，其中 A 是有效括号字符串
例如：""，"()()"，和 "()(()())" 都是有效括号字符串，嵌套深度分别为 0，1，2，而 ")(" 和 "(()" 都不是有效括号字符串。

 

给你一个有效括号字符串 seq，将其分成两个不相交的子序列 A 和 B，且 A 和 B 满足有效括号字符串的定义（注意：A.length + B.length = seq.length）。

现在，你需要从中选出 任意 一组有效括号字符串 A 和 B，使 max(depth(A), depth(B)) 的可能取值最小。

返回长度为 seq.length 答案数组 answer ，选择 A 还是 B 的编码规则是：如果 seq[i] 是 A 的一部分，那么 answer[i] = 0。否则，answer[i] = 1。即便有多个满足要求的答案存在，你也只需返回 一个。

 
* 示例 1：

> 输入：seq = "(()())"   
输出：[0,1,1,1,1,0]

* 示例 2：

>输入：seq = "()(())()"   
输出：[0,0,0,1,1,0,1,1]

* 解法一

```python

## 1、括号匹配用到栈结构，左括号入栈，右括号出栈，栈的深度就是当前字符的深度

## 2、把括号分为两个序列，就是分配两个栈，尽量保持两个栈的深度相等

## 3、根据位置的奇偶性，判断入哪个栈出哪个栈，该题的其中一种解就是 左括号和最近的右括号匹配

## 以字符串“（（（（））））”为例，先是连续的入栈，奇数位入栈 1，偶数位入栈 2，1-2-1-2；
# 当碰见出栈的字符串，改变奇偶出栈规则，偶数位出栈2，奇数位出栈1，2-1-2-1；碰见入栈的的，再改变入栈规则

## 4、利用按位与判断位置的奇偶性，用字符是否为(判断入栈的奇偶性是否改变

class Solution:
    def maxDepthAfterSplit(self, seq: str) -> List[int]:
        flg = []
        s1 = []
        s2 = []
        ans = []
        for c in seq:
            if c == '(': ## 如果 等于左括号
                if len(s1) <= len(s2): ##左栈长度小于右栈长度
                    s1.append(c)  ## 左栈 增加左括号
                    ans.append(0) ## 返回数组 该位置为A字符串 0
                    flg.append(0) ## 偶数位
                else:
                    s2.append(c)  # 右栈
                    ans.append(1)
                    flg.append(1)
            else:    ## 出现 右括号
                f = flg.pop()  ## 返回刚刚的奇偶位
                if f == 0:  ## 如果是偶数位
                    ans.append(0)  ##入左栈，返回数组该位置为0
                    s1.pop() # 去掉一个左括号
                else:
                    ans.append(1)
                    s2.pop()
        return ans
```

```python
class Solution:
    def maxDepthAfterSplit(self, seq: str) -> List[int]:
        return [i & 1 ^ (c == '(') for i, c in enumerate(seq)]
```
