#### 1. 问题描述

给你一个正整数的数组 A（其中的元素不一定完全不同），请你返回可在 一次交换（交换两数字 A[i] 和 A[j] 的位置）后得到的、按字典序排列小于 A 的最大可能排列。

如果无法这么操作，就请返回原数组。

 

* 示例 1：

>输入：[3,2,1] 
输出：[3,1,2] 
解释： 
交换 2 和 1
 

* 示例 2：

>输入：[1,1,5]     
输出：[1,1,5] 
解释：  
这已经是最小排列
 

* 示例 3：

>输入：[1,9,4,6,7]   
输出：[1,7,4,6,9]  
解释：   
交换 9 和 7
 

* 示例 4：(感觉这个示例是错的)

>输入：[3,1,1,3]  
输出：[1,3,1,3] 
解释： 
交换 1 和 3


#### 2、我的第一次解法 12.26

**2.1 思路**
        
* 先判断是不是已经是最小排列

* 按字典排序 小于 的意思是 aa <ab
* 从list末尾往前遍历，有小于A[i]的A[j]，则交换一次
* 返回交换后的值

* 卡点：遍历两次，计算量大

**2.2 代码尝试**



````python
class Solution:
    def prevPermOpt1(self, A: List[int]) -> List[int]:

        if A==sorted(A):
            return(A)
        num = 1 
        for i in range(len(A)-1, -1, -1):
            while num:
                for j in range(len(A)-1, -1, -1):
                    if A[i]<A[j]:
                        num -=1
                        c_i =i
                        c_j =j
                        break
        ai=A.pop(c_i)
        aj=A.pop(c_j)
        A.insert(c_j, ai)  
        A.insert(c_i, aj) 
        return(A)
                    
````

**2.3 正确代码解析**

问题的关键是找到交换元素的位置；

https://leetcode-cn.com/problems/previous-permutation-with-one-swap/solution/jie-ti-si-lu-by-guojian/

* 逆序遍历 找到 元素升序的位置，即A[i]<A[i-1],则A[i-1]是交换的元素之一
    
* 以i-1为终点，逆序遍历，找到交换的另一个元素A[j]，A[j]满足以下条件：



可以明确的是，第一次遍历选择出得第一次上升就保证了，第二次遍历得出的A[j]会满足以上两个条件

```python

class Solution:
    def prevPermOpt1(self, A: List[int]) -> List[int]:
        lenth = len(A)
        for i in range(lenth-1, 0, -1):
            # 第一次升序的位置，左侧元素A[i-1]大于当前元素A[i]
            if A[i-1] > A[i]:
                # 逆序从结尾向第i-1个索引查找A[j]           
                for j in range(lenth-1, i-1, -1):
                    # 元素要满足小于A[i-1]且与左侧元素不相等就是要找的最大元素
                    if A[j] < A[i-1] and A[j] != A[j-1]:
                        A[i-1],A[j] = A[j],A[i-1]
                        return A
        return A

```
